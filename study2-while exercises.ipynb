```python
# Study 2- While exercises
# study2/1 Let's print the list of numbers with while

numbers = [1,3,5,7,9,12,19,21]

i = 0
while (i < len(numbers)):
    print(numbers[i])
    i += 1
```

```python
# study2/2 - Let's take the start and end values from the user 
# and print the odd numbers on the screen

start = int(input('start : '))
finish = int(input('finish: '))

i = start
while i < finish:
    i += 1
    if (i % 2==1):
        print(i)
```

```python
# study2/4- Let's take 5 number from user and 
# sort them from smallest to largest

numbers = []

i = 0

while i<5:
    num = int(input('Number: '))
    numbers.append(num)
    i += 1

numbers.sort()

print(numbers)
```

```python
# study2/5- Let's add a product by entering the 
# product information we received from the user

products = []

number = int(input('How many products do you want to add?: '))
i = 0

while (i<number):
    name = input('Product name: ')
    price = input('Product price: ')
    products.append({
        'name' : name,
        'price': price 
    })
    i += 1
    
for product in products:
    print(f"Product name: {product['name']} product price: {product['price']}")
```
